# *Plotting of Employment Rate Data*

```python
plotdata = pd.DataFrame({
    "2019":[61.3,94.9,13.8,5.1],
    "2020":[59.5,89.7,16.2,10.3],
    "2021":[63.3,92.2,15.9,7.8],
    "2022":[64.7,94.6,14.2,5.4],
    "2023":[65.5,95.2,13.5,4.8]},
    index=["Labor_Force_Participation_Rate", "Employment_Rate", "Underemployment_Rate", "Unemployment_Rate"])
plotdata.plot(kind="bar",figsize=(15, 8))
plt.title("Employment Rate in % by Year in the Philippines")
plt.xlabel("Rate")
plt.ylabel("Ratings")
```
